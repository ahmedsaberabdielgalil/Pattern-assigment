Importing required libraries


```python
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn.naive_bayes import GaussianNB
```

Renaming coloumn

```python
# Name the columns
columns = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status",
           "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss",
           "hours-per-week", "native-country", "income"]
```

Getting training Dataset

```python
# Load training data
train_data = pd.read_csv("adult.data", names= columns, sep=', ', na_values=["?"], engine='python')
train_data
```

Getting testing Dataset

```python
# Load testing data
test_data = pd.read_csv("adult.test",skiprows=1, names= columns, sep=', ', na_values=["?"] , engine='python')
test_data
```

Removing Nulls

```python
# Drop rows with missing values from the training and testing datasets to ensure data completeness.
train_data = train_data.dropna()
test_data = test_data.dropna()
```

Data spliting

```python
# Split the data into features (X) and target variable (y)
X_train = train_data.drop("income", axis=1)
y_train = train_data["income"]
X_test = test_data.drop("income", axis=1)
y_test = test_data["income"]
```

Encoding using  One-Hot Encoding to encode the data, preparing it for training in machine learning models

```python
# Concatenate training and testing data
combined_data = pd.concat([X_train, X_test], axis=0)

# Perform one-hot encoding
combined_data_encoded = pd.get_dummies(combined_data, columns=combined_data.select_dtypes(include=["object"]).columns)

# Split back into training and testing sets
X_train_encoded = combined_data_encoded[:len(X_train)]
X_test_encoded = combined_data_encoded[len(X_train):]
```

```python
X_test_encoded
```

Filtering data more that 50k


```python
y_train_new = (y_train == '>50K').astype(int)
y_test_new = (y_test == '>50K.').astype(int)
```

instantiating Naive Bayes

```python
# Naive Bayes Classifier
model = GaussianNB()
```

Training

```python
# Train the classifier
model.fit(X_train_encoded, y_train_new)
```

Testing

```python
# Predict on test data
y_pred = model.predict(X_test_encoded)
```

model evaluation


```python
# Evaluate the classifier
tn, fp, fn, tp = confusion_matrix(y_test_new, y_pred).ravel()
# Compute Sensitivity and Specificity
sensitivity = tp / (tp + fn)
specificity = tn / (tn + fp)
print("Sensitivity (True Positive Rate):", sensitivity)
print("Specificity (True Negative Rate):", specificity)
```

```python
# Compute Posterior Probability
posterior_probs = model.predict_proba(X_test_encoded)
print("Posterior Probability of making over 50K a year:", posterior_probs[:, 1])
```
